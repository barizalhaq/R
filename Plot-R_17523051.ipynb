```r
#constant function

f<-function(x){
    fx<-4
    return(fx)
}
input<-0:10
plot(input, sapply(input,f), type="l", xlab="x", ylab="f(x)")
```

```r
#linear function

f<-function(x){
    fx<-x+5
    return(fx)
}
input<-5:20
plot(input, sapply(input,f), type="l", xlab="x", ylab="f(x)")
```

```r
#quadratic functions

f<-function(x){
    fx<-4*x^2 + 4*x + 4
    return(fx)
}
input<-1:15
plot(input, sapply(input,f), type="l", xlab="x", ylab="f(x)")
```

```r
#polynomial functions

f<-function(x){
    fx<-2*x^3 + 4*x^2 + x-2
    return(fx)
}
input<-seq(−15, 10, 0.5)
plot(input, sapply(input,f), type="l", xlab="x", ylab="f(x)")
```

```r
#rational functions

f<-function(x){
    fx<-4/2^x
    return(fx)
}
input<-seq(1, 10, 0.5)
plot(input, sapply(input,f), type="l", xlab="x", ylab="f(x)")
```
